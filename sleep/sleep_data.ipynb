```python
import numpy as np
import pandas as pd
```

```python
sleep6 = pd.read_csv('sleep-2023-06-26.csv')
sleep7 = pd.read_csv('sleep-2023-07-26.csv')
sleep8 = pd.read_csv('sleep-2023-08-25.csv')
sleep9 = pd.read_csv('sleep-2023-09-24.csv')
sleep10 = pd.read_csv('sleep-2023-10-24.csv')
sleep11 = pd.read_csv('sleep-2023-11-23.csv')
```

```python
sleep = pd.concat([sleep6, sleep7, sleep8, sleep9, sleep10, sleep11])
sleep.head()
```

### Remove unnecessary columns

```python
sleep = sleep[['dateOfSleep', 'startTime', 'endTime', 'minutesToFallAsleep', 'minutesAsleep', 'minutesAwake', 'minutesAfterWakeup', 'timeInBed', 'efficiency']]
sleep.head()
```

### Convert data types

```python
sleep['dateOfSleep'] = pd.to_datetime(sleep['dateOfSleep'])
sleep['startTime'] = pd.to_datetime(sleep['startTime']).dt.time
sleep['endTime'] = pd.to_datetime(sleep['endTime']).dt.time
sleep.head()
```

### Convert units

```python
sleep['minutesAsleep'] = sleep['minutesAsleep']/60
sleep['minutesAwake'] = sleep['minutesAwake'] /60
sleep['timeInBed'] = sleep['timeInBed'] /60
sleep.head()
```

### Rename Columns

```python
sleep.rename(columns = {'dateOfSleep':'date',
                        'startTime':'sleepStart',
                        'endTime':'sleepEnd',
                        'efficiency':'sleepEfficiency',
                        'minutesAwake':'awake',
                        'minutesAsleep':'asleep'
                        }, inplace = True) 
sleep.head()
```

### save this dataframe as sleep

```python
sleep.to_csv('sleep.csv', index=False)
```
